```python
from langgraph.graph import StateGraph,START,END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()

llm=ChatGoogleGenerativeAI(model="gemini-2.0-flash")
```

```python
class llmState(TypedDict):
    title:str
    outline:str
    content:str
```

```python
def gen_outline(state:llmState)->llmState:

    title=state['title']
    promt=f'Generate a outline on the topic - {title}'

    outline=llm.invoke(promt).content
    print(outline)
    state['outline']=str(outline)

    return state
```

```python
def gen_content(state:llmState)->llmState:

    title=state['title']
    outline=state['outline']
    prompt=f'wite a detailed blog on the title -{title} using the following outline \n {outline}'

    content=llm.invoke(prompt).content
    state['content']=str(content)

    return state
```

```python
graph=StateGraph(llmState)

graph.add_node('gen_outline',gen_outline)
graph.add_node('gen_content',gen_content)

graph.add_edge(START,'gen_outline')
graph.add_edge('gen_outline','gen_content')
graph.add_edge('gen_content',END)

workflow=graph.compile()
```

```python
Intial_state={'title':'Rise of AI in india'}
Final_state=workflow.invoke(Intial_state)
print(Final_state)
```
